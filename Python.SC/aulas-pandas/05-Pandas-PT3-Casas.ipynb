```python
import pandas as pd
import re
```

```python
df = pd.read_csv('K/house.csv/house.csv')
df
```

```python
## AGRUPANDO VALORES NUMERICOS ##
```

```python
 ## MEDIA DOS VALORES NUMERICOS ##
grupo1 = df.groupby(['SellerG']).mean(numeric_only=True)
grupo1
```

```python
## SOMA E PRIORIDADE por chave E QUANTIDADE Q DEVE APARECER ##
grupo1 = df.groupby(['SellerG']).sum(numeric_only=True).sort_values('Rooms', ascending = False).head(10)
grupo1
```

```python
## SOMA DE TODOS OS VALORES POSSIVEIS ##
grupo1 = df.groupby(['SellerG']).count().sort_values('Rooms', ascending = False).head(10)
grupo1
```
